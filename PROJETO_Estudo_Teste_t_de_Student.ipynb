```python
import numpy as np
from scipy.stats import ttest_ind

# Dataset fornecido
pesos_operadores = {
    'A': [876, 872, 848, 866, 887, 870, 865, 860, 811, 869, 862, 873, 875, 845, 874, 859, 884, 877, 856, 842, 859, 865, 864, 877, 881, 842, 877, 856, 829, 875, 855, 861, 863, 879, 858, 892, 851, 868, 872, 881, 865, 864, 867],
    'B': [834, 810, 838, 818, 1018, 811, 825, 798, 837, 827, 792, 805, 848, 798, 837, 827, 815, 827, 815, 826, 855, 826, 855, 861, 863, 879, 858, 892, 851, 868, 872, 881, 865, 864, 819, 823, 809, 813, 812, 803, 806, 795, 835, 831, 824]
}

# Realize o teste t de Student para amostras independentes
t_statistic, p_value = ttest_ind(pesos_operadores['A'], pesos_operadores['B'])

# Imprima os resultados
print(f'Estatística t: {t_statistic}')
print(f'Valor-p: {p_value}')

# Verifique se rejeitamos a hipótese nula (significância de 0.05)
alpha = 0.05
if p_value < alpha:
    print('Rejeitamos a hipótese nula: as médias dos pesos dos operadores A e B são diferentes.')
else:
    print('Não há evidências suficientes para rejeitar a hipótese nula: as médias são semelhantes.')
```

```python
import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt

# Dataset fornecido
pesos_operadores = {
    'A': [876, 872, 848, 866, 887, 870, 865, 860, 811, 869, 862, 873, 875, 845, 874, 859, 884, 877, 856, 842, 859, 865, 864, 877, 881, 842, 877, 856, 829, 875, 855, 861, 863, 879, 858, 892, 851, 868, 872, 881, 865, 864, 867],
    'B': [834, 810, 838, 818, 1018, 811, 825, 798, 837, 827, 792, 805, 848, 798, 837, 827, 815, 827, 815, 826, 855, 826, 855, 861, 863, 879, 858, 892, 851, 868, 872, 881, 865, 864, 819, 823, 809, 813, 812, 803, 806, 795, 835, 831, 824]
}

# Realize o teste t de Student para amostras independentes
t_statistic, p_value = ttest_ind(pesos_operadores['A'], pesos_operadores['B'])

# Imprima os resultados
print(f'Estatística t: {t_statistic}')
print(f'Valor-p: {p_value}')

# Configuração do layout
fig, ax = plt.subplots(figsize=(8, 6))

# Plotando barras de erro para as médias dos operadores A e B
means = [np.mean(pesos_operadores['A']), np.mean(pesos_operadores['B'])]
std_devs = [np.std(pesos_operadores['A']), np.std(pesos_operadores['B'])]
bar_width = 0.35
bar_positions = np.arange(2)

ax.bar(bar_positions, means, bar_width, yerr=std_devs, capsize=5, alpha=0.7, align='center', label=['Operador A', 'Operador B'])

# Adicionando rótulos e legendas
ax.set_xticks(bar_positions)
ax.set_xticklabels(['Operador A', 'Operador B'])
ax.set_ylabel('Peso (mg)')
ax.set_title('Comparação de Pesos entre Operadores A e B')
ax.legend()

# Exibindo a estatística t e valor-p no gráfico
textstr = f'Estatística t = {t_statistic:.2f}\nValor-p = {p_value:.4f}'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=10, verticalalignment='top', bbox=props)

plt.show()
```

```python
import numpy as np
from scipy.stats import ks_2samp

# Gere duas amostras, por exemplo, de distribuições normais
np.random.seed(42)
sample1 = np.random.normal(0, 1, 1000)
sample2 = np.random.normal(0, 1.2, 1200)

# Realize o teste de Kolmogorov-Smirnov
statistic, p_value = ks_2samp(sample1, sample2)

# Imprima os resultados
print(f'Estatística de teste: {statistic}')
print(f'Valor-p: {p_value}')

# Verifique se rejeitamos a hipótese nula (significância de 0.05)
alpha = 0.05
if p_value < alpha:
    print('Rejeitamos a hipótese nula: as distribuições são diferentes.')
else:
    print('Não há evidências suficientes para rejeitar a hipótese nula.')
```

```python
import numpy as np
from scipy.stats import ks_2samp
import matplotlib.pyplot as plt

# Gere duas amostras, por exemplo, de distribuições normais
np.random.seed(42)
sample1 = np.random.normal(0, 1, 1000)
sample2 = np.random.normal(0, 1.2, 1200)

# Realize o teste de Kolmogorov-Smirnov
statistic, p_value = ks_2samp(sample1, sample2)

# Imprima os resultados
print(f'Estatística de teste: {statistic}')
print(f'Valor-p: {p_value}')

# Verifique se rejeitamos a hipótese nula (significância de 0.05)
alpha = 0.05
if p_value < alpha:
    print('Rejeitamos a hipótese nula: as distribuições são diferentes.')
else:
    print('Não há evidências suficientes para rejeitar a hipótese nula.')

# Plotando histogramas das amostras
plt.hist(sample1, bins=30, density=True, alpha=0.5, label='Amostra 1')
plt.hist(sample2, bins=30, density=True, alpha=0.5, label='Amostra 2')

# Plotando a CDF empírica
ecdf_sample1 = np.arange(1, len(sample1) + 1) / len(sample1)
ecdf_sample2 = np.arange(1, len(sample2) + 1) / len(sample2)

plt.step(np.sort(sample1), ecdf_sample1, label='CDF Empírica - Amostra 1', linestyle='--')
plt.step(np.sort(sample2), ecdf_sample2, label='CDF Empírica - Amostra 2', linestyle='--')

plt.title('Distribuição e CDF Empírica')
plt.xlabel('Valores')
plt.ylabel('Densidade e Probabilidade Cumulativa')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.stats import ks_2samp
import matplotlib.pyplot as plt

# Gere duas amostras, por exemplo, de distribuições normais
np.random.seed(42)
sample1 = np.random.normal(0, 1, 1000)
sample2 = np.random.normal(0, 1.2, 1200)

# Realize o teste de Kolmogorov-Smirnov
statistic, p_value = ks_2samp(sample1, sample2)

# Imprima os resultados
print(f'Estatística de teste: {statistic}')
print(f'Valor-p: {p_value}')

# Verifique se rejeitamos a hipótese nula (significância de 0.05)
alpha = 0.05
if p_value < alpha:
    print('Rejeitamos a hipótese nula: as distribuições são diferentes.')
else:
    print('Não há evidências suficientes para rejeitar a hipótese nula.')

# Configuração do layout
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Plotando histogramas das amostras
axes[0].hist(sample1, bins=30, density=True, alpha=0.5, label='Amostra 1')
axes[0].hist(sample2, bins=30, density=True, alpha=0.5, label='Amostra 2')
axes[0].set_title('Histograma')
axes[0].set_xlabel('Valores')
axes[0].set_ylabel('Densidade')
axes[0].legend()

# Plotando a CDF empírica
ecdf_sample1 = np.arange(1, len(sample1) + 1) / len(sample1)
ecdf_sample2 = np.arange(1, len(sample2) + 1) / len(sample2)

axes[1].step(np.sort(sample1), ecdf_sample1, label='CDF Empírica - Amostra 1', linestyle='--')
axes[1].step(np.sort(sample2), ecdf_sample2, label='CDF Empírica - Amostra 2', linestyle='--')
axes[1].set_title('Curva CDF Empírica')
axes[1].set_xlabel('Valores')
axes[1].set_ylabel('Probabilidade Cumulativa')
axes[1].legend()

plt.tight_layout()
plt.show()
```

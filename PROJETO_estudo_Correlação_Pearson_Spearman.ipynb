```python
import pandas as pd
import seaborn as sns

# Carregar o conjunto de dados Heart Disease UCI
heart_data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data", header=None, na_values="?")

# Adicionar nomes de colunas
column_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
heart_data.columns = column_names

# Exibir as primeiras linhas do conjunto de dados
print(heart_data.head())

# Calcular correlação de Pearson
pearson_corr = heart_data['age'].corr(heart_data['trestbps'], method='pearson')

# Calcular correlação de Spearman
spearman_corr = heart_data['age'].corr(heart_data['trestbps'], method='spearman')

# Imprimir resultados
print(f"Correlação de Pearson: {pearson_corr:.4f}")
print(f"Correlação de Spearman: {spearman_corr:.4f}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o conjunto de dados Heart Disease UCI
heart_data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data", header=None, na_values="?")

# Adicionar nomes de colunas
column_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
heart_data.columns = column_names

# Exibir as primeiras linhas do conjunto de dados
print(heart_data.head())

# Calcular correlação de Pearson
pearson_corr = heart_data['age'].corr(heart_data['trestbps'], method='pearson')

# Calcular correlação de Spearman
spearman_corr = heart_data['age'].corr(heart_data['trestbps'], method='spearman')

# Imprimir resultados
print(f"Correlação de Pearson: {pearson_corr:.4f}")
print(f"Correlação de Spearman: {spearman_corr:.4f}")

# Criar um DataFrame de exemplo para o gráfico de dispersão
data = {
    'Variavel1': heart_data['age'],
    'Variavel2': heart_data['trestbps']
}

df = pd.DataFrame(data)

# Configurar a estética do seaborn
sns.set(font_scale=1.2)
sns.set_style('whitegrid')

# Criar os gráficos de dispersão
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x='Variavel1', y='Variavel2', data=df)
plt.title(f'Correlação de Pearson: {pearson_corr:.4f}')

plt.subplot(1, 2, 2)
sns.scatterplot(x='Variavel1', y='Variavel2', data=df)
plt.title(f'Correlação de Spearman: {spearman_corr:.4f}')

plt.tight_layout()
plt.show()
```
